```python
import pandas as pd
from tqdm.auto import tqdm

file_path = '/content/drive/MyDrive/laliga_0158.csv'
df = pd.read_csv(file_path)
```

```python
news_lst = []
for n in tqdm(range(len(df))):
    pos = len(df)-n-1
    news_sen = pipe(df['text'].iloc[pos])[0]['summary_text']
    time_sen = df['time'].iloc[pos]
    news_lst.append('第' + time_sen + '分钟，'+ news_sen + '。')
```
